```python
data = pd.read_csv('samplesize.csv')
data1 = pd.DataFrame(data)
chart = alt.Chart(data)
data2= data1.loc[0:55]
```

```python
data1.columns
```

```python
brush = alt.selection_interval
```

```python
alt.Chart(data1).mark_circle().encode(
    alt.X('Positive affect', title="Positive affect",scale=alt.Scale(zero=False)),
    alt.Y('Log GDP per capita',title="GDP", scale=alt.Scale(zero=False, padding=1)),
    color='Country name',
    size='Life Ladder',
    tooltip=[alt.Tooltip('Positive affect'),
            alt.Tooltip('Log GDP per capita'),
            alt.Tooltip('Country name'),
            alt.Tooltip('Life Ladder')]
).properties(width=500,height=400,title="Positivity vs GDP and its link to happines")
```

```python
alt.Chart(data1).mark_line().encode(
    alt.X('Country name',title="Country"),
    alt.Y('Positive affect'),
    tooltip=['Country name','Positive affect']
).properties(title="World's Happines Index",width=600,height=400)
```

```python
alt.Chart(data1).mark_line().encode(
    alt.X('Country name',title="Country"),
    alt.Y('Negative affect'),
    tooltip=['Country name','Negative affect']
).properties(title="World's Happines Index",width=600,height=400)
```

```python
alt.Chart(data1).mark_circle().encode(
    alt.X('Positive affect', title="Positive affect",scale=alt.Scale(zero=False)),
    alt.Y('Freedom to make life choices',title="Freedom to make life choices", scale=alt.Scale(zero=False, padding=1)),
    color='Country name',
    size='Life Ladder',
    tooltip=[alt.Tooltip('Positive affect'),
            alt.Tooltip('Log GDP per capita'),
            alt.Tooltip('Country name'),
            alt.Tooltip('Life Ladder')]
).properties(width=500,height=400,title="Freedom and its link to happines")
```

```python
line = alt.Chart(data1).mark_line(
    color='blue',
    size=3
).transform_window(
    rolling_mean='mean(Close)',
    frame=[-15, 15]
).encode(
    x='Country name',
    y='Positive affect'
) 

points = alt.Chart(data1).mark_point().encode(
    x='Country name',
    y=alt.Y('Positive affect',axis=alt.Axis(title='Positive affect')),
    tooltip=[alt.Tooltip('Positive affect'),
            alt.Tooltip('Country name')]
).properties(title="Freedom and its link to happines",width=600,height=400)

points + line
```
